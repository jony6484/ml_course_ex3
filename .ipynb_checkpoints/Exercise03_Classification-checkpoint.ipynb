```python
lFullName = ['Age', 'Sex', 'ChestPainType', 'RestingBloodPressure', 'Cholesterol', 'FastingBloodSugar', 'RestECG', 'MaxHeartRateAchieved', 'ExerciseInducedAngina', 'StDepression', 'StSlope', 'NumMajorVessels', 'Thalassemia', 'Target']
space     = max(map(len, dData.columns))
for (shortName, fullName) in zip(dData.columns, lFullName):
    print(f'{shortName:{space}s} = {fullName}')
```

### Get data and normalization:

```python
mX = dData.drop(columns='condition').values
vY = dData['condition'].values

mX.shape, vY.shape
```

```python
#-- Normalize data:
mX -= mX.mean(0)
mX /= mX.std(0)
```

### Basic classification:
Let us try several default classifiers

```python
from sklearn.linear_model    import LogisticRegression
from sklearn.svm             import SVC
from sklearn.neighbors       import KNeighborsClassifier
from sklearn.tree            import DecisionTreeClassifier
from sklearn.model_selection import cross_val_predict, KFold

#-- Classifiers:
#-- You can set the hyper-parameters as well
lModels  = []
lModels += [(SVC                   (C=1), 'SVM'                )]
lModels += [(LogisticRegression    (),    'Logistic Regression')]
lModels += [(KNeighborsClassifier  (),    'KNN'                )]
lModels += [(DecisionTreeClassifier(),    'Tree'               )]


N = len(vY)
for (oClassifier, name) in lModels:
    vHatY    = cross_val_predict(oClassifier, mX, vY, cv=KFold(N))
    accuracy = np.mean(vY == vHatY)
    print(f'{name:19s} = {100*accuracy:2.2f}%')
```

### Exercise A:
* Use `KNeighborsClassifier` and get above 84% leave-one-out cross validation accuracy.  
(Play with the hyper-parameters)

# My Solution 1

```python
cv_accuracy_results = []
n_neighbors = np.arange(1,51)
for nn in n_neighbors:
    vHatY = cross_val_predict(KNeighborsClassifier(n_neighbors=nn), mX, vY, cv=KFold(N))
    cv_accuracy_results.append(np.average(vHatY == vY))
```

```python
cv_knn_df = pd.DataFrame({'Neighbors' : n_neighbors, 'Accuracy': cv_accuracy_results})
cv_knn_df.sort_values(by=["Accuracy"], inplace=True, ascending=False)
cv_knn_df.head(10)
# Train Test split???? We just cross validated, why not split te data before cv?
```

```python
plt.bar(cv_knn_df["Neighbors"],cv_knn_df["Accuracy"]*100, edgecolor='k')
plt.xlabel('# Neighbors')
plt.ylabel('Accuracy %')
plt.ylim([75,85])
plt.xlim([0,51])
plt.title('KNN Cross validation of N Neighbors')
```

### Exercise B:
1. Use `dData3` and get above 85% leave-one-out cross validation accuracy.
2. You are allowed you to only 6 features (from the given 18).  
Get more than 80% accuracy with only 6 features.  
What is the right approach for this task?
**Extra:** Get above 85.5% (using only 6 features).
---

```python
def make_sub_mat(best_features, feature, df):
    
    if len(best_features) == 0:
        mX_sub = df[feature].values[:, None]
    else:
        mX_sub = df[best_features].values
        mX_sub = np.hstack([mX_sub, dData3[feature].values[:, None]])
    
    mX_sub = (mX_sub - mX_sub.mean(0)) / mX_sub.std(0)    
    return mX_sub    
    
```

```python
# Gridi way of finding the best n features
def find_best_n_feats(oClassifier, df, n, kf=0):
    
      
    features = np.array(df.columns)
    target = features[-1]
    features = features[:-1]
    
    feats_accu_df = pd.DataFrame(index=features)

    vY = df[target]
    N = len(vY)
    if kf==0:
        kf = N

    best_features = np.array([])

    rem_features = features.copy()
    n=6
    for i in range(n):
        df_col_name = str(i+1) + ' features'
        for feature in rem_features:
            mX_sub = make_sub_mat(best_features, feature, df)
            vYhat_sub = cross_val_predict(oClassifier, mX_sub, vY, cv=KFold(kf), n_jobs=-1)
            accu = np.mean(vYhat_sub == vY)
            feats_accu_df.loc[feature, df_col_name] = accu
        best_i_feature = feats_accu_df[df_col_name].argmax()
        best_i_feature_name = features[best_i_feature]
        best_features = np.append(best_features, best_i_feature_name)
        ind = np.argwhere(rem_features==best_i_feature_name)
        rem_features = np.delete(rem_features, ind)
        
    return best_features, feats_accu_df.iloc[:,-1].max()
```

```python
# Chossing the best model & features
lModels  = []
lModels += [(SVC                   (C=10)               , 'SVM'                )]
lModels += [(LogisticRegression    (fit_intercept=False), 'Logistic Regression')]
lModels += [(KNeighborsClassifier  ()                   , 'KNN'                )]
lModels += [(DecisionTreeClassifier()                   , 'Tree'               )]


N = len(vY)
for (oClassifier, name) in lModels:
    best_features, accuracy = find_best_n_feats(oClassifier, dData3, 6, kf=N)
    print(f'{name:19s} = {100*accuracy:2.2f}%')
```

```python
# My solution:
# ^^^ The best classifier (above 85%): Logistic Regression ^^^
# Re running for specific results:
best_features, accuracy = find_best_n_feats(LogisticRegression(fit_intercept=False), dData3, 6, kf=20)
print(f'Accuracy - logistic regression: {accuracy*100:2.2f}%')
print(f'Selected features - logistic regression: {best_features}')
```

### Grid Search:

```python
from sklearn.model_selection import GridSearchCV

dSvmParams = {'kernel': ('linear',),
              'C'     : np.linspace(0.01, 10, 21)}
oSearch    = GridSearchCV(SVC(), dSvmParams, cv=KFold(N, shuffle=True)).fit(mX3, vY)

print(f'Linear best parameters CV score = {oSearch.best_score_}')
print(oSearch.best_params_)
```
